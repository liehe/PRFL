```python
from src.kmeans import KMeans
import functools

class ThresholdingV2(object):
    def __init__(self, K, num_thresholding_steps, num_sampled_devices,
                 strategy, rng):
        self.num_thresholding_steps = num_thresholding_steps
        self.num_sampled_devices = num_sampled_devices
        self.strategy = strategy
        self.rng = rng

        self.tau_fn = self.get_tau_fn(strategy, K)

    def get_tau_fn(self, strategy, K):
        K = 3
        if strategy == 'adaptive-1':
            simple_clustering_oracle = KMeans(K=K, iterations=10)

            def fn(c, a):
                c = sorted(c)
                return (c[1] + c[2])/2 
            return functools.partial(
                self._adaptive_radius,
                clustering_oracle=simple_clustering_oracle,
                fn=fn)

        if strategy == 'adaptive-2':
            simple_clustering_oracle = KMeans(K=K, iterations=10)

            def fn(c, a):
                c = sorted(c)
                return c[1] * 0.99
            return functools.partial(
                self._adaptive_radius,
                clustering_oracle=simple_clustering_oracle,
                fn=fn)

        if strategy == 'adaptive-3':
            simple_clustering_oracle = KMeans(K=K, iterations=10)

            def fn(c, a):
                sorted_c = sorted(list(zip(range(len(c)), c)),
                                  key=lambda x: x[1])
                index_c, min_c = sorted_c[0]
                s = sum(i == index_c for i in a)
                if s <= 2:
                    return (sorted_c[1][1] + sorted_c[2][1]) / 2
                else:
                    return (sorted_c[0][1] + sorted_c[1][1]) / 2
            return functools.partial(
                self._adaptive_radius,
                clustering_oracle=simple_clustering_oracle,
                fn=fn)

        if strategy == 'adaptive-4':
            simple_clustering_oracle = KMeans(K=K, iterations=10)

            def fn(c, a):
                c = sorted(c)
                return (c[1] + c[0])/2 
            return functools.partial(
                self._adaptive_radius,
                clustering_oracle=simple_clustering_oracle,
                fn=fn)

        if strategy == 'adaptive-5':
            simple_clustering_oracle = KMeans(K=K, iterations=10)

            def fn(c, a):
                c = sorted(c)
                return c[1] * 0.9 + c[0] * 0.1
            return functools.partial(
                self._adaptive_radius,
                clustering_oracle=simple_clustering_oracle,
                fn=fn)

        if strategy.startswith('Q'):
            quantile = float(strategy[1:])
            assert quantile <= 1 and quantile >= 0
            return functools.partial(
                self._quantile_radius, quantile=quantile
            )

        raise NotImplementedError(strategy)

    def __call__(self, momentums, center0):
        total_devices = len(momentums)

        center = center0.clone()
        centers = []
        assignments = []
        for l in range(self.num_thresholding_steps):
            if self.num_sampled_devices == 'all':
                indices = np.arange(total_devices)
            else:
                indices = self.rng.choice(
                    np.arange(total_devices),
                    size=self.num_sampled_devices,
                    replace=False)
            sampled_momentum = [momentums[i] for i in indices]
            tau = self.tau_fn(center, sampled_momentum)

            ys = []
            i_s = []
            for i in indices:
                m = momentums[i]
                if torch.linalg.norm(m - center) < tau:
                    y = m
                    i_s.append(i)
                else:
                    y = center
                ys.append(y)
            center = sum(ys) / len(ys)
            centers.append(center.clone())
            assignments.append(i_s)
#             print(len(i_s), i_s)

        return centers, assignments

    def _adaptive_radius(self, v, momentums, clustering_oracle, fn):
        distances = torch.Tensor([torch.linalg.norm(v - m) for m in momentums])
        distances = torch.unsqueeze(distances, dim=1)
        a, c = clustering_oracle(distances)
        while len(c) < 2:
            a, c = clustering_oracle(distances)
        return fn(c, a)

    def _quantile_radius(self, v, momentums, quantile):
        distances = sorted([torch.linalg.norm(v - m) for m in momentums])
#         print(v.shape, len(momentums), quantile, distances, int(len(distances) * quantile))
        return distances[int(len(distances) * quantile)]
```

```python
import numpy as np
import torch
from tqdm import tqdm
def _initialize_cluster_center_perfect(y, K):
    y = pd.Series(y)
    indices = []
    for k in range(K):
        index = y[y == k].index
        indices.append(np.random.choice(index))
    return indices

L = 50
# data = []
# for std, X, y in [(1, X1, y1), (2, X2, y2), (0.5, X3, y3)]:
for std, X, y in [(4, X4, y4)]:
    X = torch.Tensor(X)
    for seed in tqdm(range(10)):
        rng = np.random.default_rng(seed)
        initial_center_indices = _initialize_cluster_center_perfect(y, K)

    #     for strategy in ["adaptive-1", "adaptive-2", "adaptive-3", "Q0.09", "Q0.11"]:
    #     for strategy in ["Q0.05", "Q0.2"]:
        for strategy in ["Q0.1"]:
            clustering_oracle = ThresholdingV2(K=K, rng=rng, num_thresholding_steps=L, 
                                               num_sampled_devices='all', strategy=strategy)

            cluster_centers = []
            initial_centers = [X[i] for i in initial_center_indices]

            for i in range(K):
                groundtruth_cluster_id = y[initial_center_indices[i]]
                groundtruth_assignment = np.arange(len(y))[np.where(y == groundtruth_cluster_id, True, False)]
                centers, assignments = clustering_oracle(X, initial_centers[i])
                cluster_centers.append(centers)

                def _compute_distance(center, X, y, groundtruth_assignment):
                    distance = []
                    for assignment in groundtruth_assignment:
                        distance.append(np.linalg.norm(X[assignment] - center) ** 2)
                    return np.sqrt(sum(distance) / len(distance))

                for l, (center, assignment) in enumerate(zip(centers, assignments)):
#                     jaccard_distance = len(set(groundtruth_assignment).intersection(set(assignment))) / len(set(groundtruth_assignment).union(set(assignment)))
                    distance = _compute_distance(center, X, y, groundtruth_assignment)
                    data.append({
                        "seed": seed, "iteration": l, "distance": distance, "k": i,
                        "strategy": strategy, "std": std
                    })
```

```python
def compute_jaccard_distance(prediction, groundtruth):
    df = pd.DataFrame({"prediction": prediction, "groundtruth": groundtruth})

    def find_mapping_udf(group):
        predicted_groups = group["prediction"].value_counts(ascending=False)
        return predicted_groups.index[0]

    # groundtruth cluster id => predicted cluster id
    mapping = df.groupby("groundtruth").apply(find_mapping_udf).to_dict()

    def intersection_udf(group):
        predicted_groups = group["prediction"].value_counts(ascending=False)
        return predicted_groups.iloc[0]

    intersection = df.groupby("groundtruth").apply(intersection_udf)

    def union_udf(group):
        predicted_groups = group["prediction"].value_counts(ascending=False)
        # We assume the largest value in the prediction is the class. Then its count is the size of intersection.
        intersection = predicted_groups.iloc[0]
        return intersection

    union = []
    for k in range(len(mapping)):
        union.append(len(df[(df["prediction"] == mapping[k]) | (df["groundtruth"] == k)]))
    union = pd.Series(union)
    return (intersection / union).mean()
```

```python
def predict(X, y, X_pred, degree=3):
    """
    Predict iteration l with the given distance.
    """
    p = np.polyfit(X, y, degree)
    return np.poly1d(p)(X_pred)

def predict_contour(a, level):
    df = pd.DataFrame({"z": (a - a.shift(1))/(a.shift(1)-a.min()+0.001) , "x": np.arange(len(a)), "y": a})
    df = df.dropna()
    print(df['z'])
    x = predict(df["z"], df["x"], level)
    y = predict(df["x"], df["y"], x)
    return x, y
    

# for level in [-0.05, -0.04, -0.03, -0.0]
df1 = df.groupby(["std", "iteration"]).mean()

contour = []
# for level in [-0.01, -0.02, -0.03, -0.04]:
for level in [-0.19]:
    contour_x1 = []
    contour_y1 = []
    for std in [0.5, 1, 2, 4]:
        a = df1.loc[std]['distance']
        _x, _y = predict_contour(a, level)
        contour_x1.append(_x)
        contour_y1.append(_y)
    
    _cx = np.arange(1, 20)
    _cy = predict([0]+contour_x1, [0]+contour_y1, _cx, degree=4)
    contour.append((_cx, _cy))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
font = {"size": 10}
plt.rc("font", **font)

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
plt.style.use("tableau-colorblind10")

fig, ax = plt.subplots(1, 1, figsize=(5, 4))
ax.text(10, 19.5, '"Elbow" of the curves')
for contour_x1, contour_y1 in contour:
    ax.plot(contour_x1, contour_y1, "--", color='grey')
_df = df[df["strategy"].apply(lambda x: x.startswith("Q"))].copy()
_df = _df.sort_values("std", ascending=False)
_df[r"Strategy $\tau$"] = _df["strategy"].apply(lambda x: "Quantile {}%".format(float(x[1:])*100))
_df["Average distance to cluster center"] = _df[r"distance"]
_df[r"Number of clipping iterations ($\ell$)"] = _df["iteration"] 
_df["std"] = _df["std"].apply(lambda x: r"$\sigma=$"+str(x)) 
g = sns.lineplot(
    data=_df,
    x=r"Number of clipping iterations ($\ell$)",
    y="Average distance to cluster center",
    hue=r"std",
    ax=ax
)
ax.set_xlim(0, 50)
ax.set_ylim(0, 20)
```
